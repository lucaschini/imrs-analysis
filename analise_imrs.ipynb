ESGOTO

```python
df_esgoto = df.drop([
       'ibge7','tx_mort_acid_15_29', 'tx_mort_hom_15_29', 'tx_mort_mama',
       'tx_mort_inf', 'existe_conselho_saude', 'tx_analf_25m', 'prop_fund_25m',
       'perc_med_25m', 'tx_med_adeq', 'tx_med_lab', 'existe_conselho_educ',
       'tx_crimes_pat', 'tx_crimes_pes', 'tx_hom', 'existe_uni_infra',
       'perc_energia', 'perc_banagua', 'perc_lixo', 'pop_total',
       'pop_masc', 'pop_fem', 'raz_dep', 'tx_urb', 'tx_fec', 'esp_vida',
       'rdpc', 'renocup', 'perc_rdpc_140', 'fam_pbf', 'tx_emp_form',
       'rend_pc_form', 'pib_pc', 'perc_agro', 'perc_serv', 'tx_ativ_18m'], axis=1)
df_esgoto
```

Começando o trabalho com uma visão geral do tratamento de esgoto em Minas Gerais


```python
bw = 2*(75.362500-37.492500)/np.cbrt(len(df_esgoto))
bs = np.ceil((98.220000-0.03000)/bw)
sns.displot(data=df_esgoto, x=df_esgoto.perc_esgoto, bins=bs, binwidth=bw)
plt.title("Cobertura de Rede de Esgoto nos Municípios de MG: Quantos Têm Acesso?", pad=20)
plt.xlabel("Percentual de Domicílios com Rede de Esgoto (%)")
plt.ylabel("Número de Municípios")
plt.show()
```

Da pra ver que a maioria dos municipios tem uma boa quantidade de acesso ao esgoto, mas tem mais de 60 municipios que tem menos de 10% de acesso a esgoto

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df_esgoto, y='perc_esgoto')
plt.title("Cobertura de Rede de Esgoto nos Municípios de MG: Métricas")
plt.ylabel("Percentual de Domicílios com Rede de Esgoto (%)")
plt.xlabel("Minas Gerais")
plt.show()
```

AGUA


perc_banagua -> refere-se ao "percentual de domicílios com banheiro e água encanada". Este indicador é utilizado para avaliar as condições de saneamento básico nos municípios, refletindo a proporção de residências que possuem acesso a instalações sanitárias adequadas e abastecimento de água potável.




```python
df_banagua = df.drop([
       'ibge7','tx_mort_acid_15_29', 'tx_mort_hom_15_29', 'tx_mort_mama',
       'tx_mort_inf', 'existe_conselho_saude', 'tx_analf_25m', 'prop_fund_25m',
       'perc_med_25m', 'tx_med_adeq', 'tx_med_lab', 'existe_conselho_educ',
       'tx_crimes_pat', 'tx_crimes_pes', 'tx_hom', 'existe_uni_infra',
       'perc_energia', 'perc_esgoto', 'perc_lixo', 'pop_total',
       'pop_masc', 'pop_fem', 'raz_dep', 'tx_urb', 'tx_fec', 'esp_vida',
       'rdpc', 'renocup', 'perc_rdpc_140', 'fam_pbf', 'tx_emp_form',
       'rend_pc_form', 'pib_pc', 'perc_agro', 'perc_serv', 'tx_ativ_18m'], axis=1)
df_banagua
```

```python
bw1 = 2*(99.000000-38.310000)/np.cbrt(len(df_banagua))
bs1 = np.ceil((100.000000-38.310000)/bw1)
sns.displot(data=df_banagua, x=df_banagua.perc_banagua, bins=bs1, binwidth=bw1)
plt.title("Cobertura de Banheiro e Água Encanada nos Municípios de Minas: Um Panorama")
plt.xlabel("Percentual de domicílios com banheiro e água encanada (%)")
plt.ylabel("Número de municípios")
plt.show()
```

```python
df_banagua["perc_banagua"].describe()
```

```python
plt.figure(figsize=(8,8)) # ajustando figura
sns.boxplot(data=df_banagua, y='perc_banagua')
plt.title("Domicílios com banheiro e água encanada em MG: Métricas")
plt.ylabel("Percentual de domicílios com banheiro e água encanada em MG (%)")
plt.xlabel("Minas Gerais")
plt.show()
```
